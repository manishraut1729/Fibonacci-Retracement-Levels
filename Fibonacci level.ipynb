# Fibonacci Retracement Levels-Technical Indicator for Stock Analysis
# 


# Importing Important Library

```python
import pandas_datareader as pdr
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Importing DataSet from Yahoo Finance

```python
df = pdr.get_data_yahoo('AXISBANK.NS', 
                          start=dt.datetime(2020, 11, 1), 
                          end=dt.datetime.now())
df
```

# Finding all Fibonacci Retracement Levels

```python
maxp=df['Close'].max()
minp=df['Close'].min()
diff=maxp-minp
level_1=maxp-diff*0.236
level_2=maxp-diff*0.382
level_3=maxp-diff*0.5
level_4=maxp-diff*0.618
```

# Printing all Fibonacci Retracement Levels

```python
print('Level_%\t Price(Rs)')
print('0.0%\t', maxp)
print('23.6%\t', level_1)
print('38.2%\t', level_2)
print('50.0%\t', level_3)
print('61.8%\t', level_4)
print('100%\t', minp)
```

# Plotting the Fibonacci Retracement Levels

```python
plt.figure(figsize=(12.33,5))
plt.title('Fibonacci Level')
plt.plot(df.index, df['Close'])
plt.axhline(maxp, linestyle='-', alpha=0.5, color='red')
plt.axhline(level_1, linestyle='--', alpha=0.5, color='green')
plt.axhline(level_2, linestyle='--', alpha=0.5, color='blue')
plt.axhline(level_3, linestyle='--', alpha=0.5, color='blue')
plt.axhline(level_4, linestyle='--', alpha=0.5, color='green')
plt.axhline(minp, linestyle='-', alpha=0.5, color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

# Filling the colour between the levels

```python
fig=plt.figure(figsize=(12.33,5))
ax=fig.add_subplot(111)
plt.title('Fibonacci Level')
plt.plot(df.index, df['Close'])
plt.axhline(maxp, linestyle='-', alpha=0.5, color='red')
plt.axhline(level_1, linestyle='--', alpha=0.5, color='green')
plt.axhline(level_2, linestyle='--', alpha=0.5, color='blue')
plt.axhline(level_3, linestyle='--', alpha=0.5, color='blue')
plt.axhline(level_4, linestyle='--', alpha=0.5, color='green')
plt.axhline(minp, linestyle='-', alpha=0.5, color='red')
ax.fill_between(df.index, maxp, level_1, color='yellow')
ax.fill_between(df.index, level_1, level_2, color='green')
ax.fill_between(df.index, level_2, level_3, color='white')
ax.fill_between(df.index, level_3, level_4, color='green')
ax.fill_between(df.index, level_4, minp, color='yellow')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
